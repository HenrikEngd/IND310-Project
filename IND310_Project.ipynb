```python
# Extracting stock data for selected companies
tickers = ['EQNR.OL', 'DNB.OL', 'AKRBP.OL', 'ORK.OL', 'MOWI.OL']
data = yf.download(tickers)

data.head()
```

```python
data_reset = data.reset_index()

# Print the columns of the data to verify
print(data_reset.columns)

# Select only the 'Date' and 'Close' columns for a specific ticker (e.g., EQNR.OL)
data_close = data_reset[[('Date', ''), ('Close', 'EQNR.OL')]].copy()
# Flatten the column names
data_close.columns = ['Date', 'Close']

# Print the head of the new DataFrame to verify
print(data_close.head())
```

```python
data_close.rename(columns={'Date' : 'date', 'Close' : 'price_t'}, inplace=True)
data_close.head()
```

```python
# Shift the price_t column by 1 index to obtain price_t-1
data_close['price_t-1'] = data_close['price_t'].shift(1)
data_close.head()
```

```python
data_close['returns'] = (data_close['price_t'] / data_close['price_t-1']) - 1
data_close.head()
```

```python
ax = data_close['price_t'].plot(figsize=(12, 8))
ax.set_xlabel('Date')
ax.set_ylabel('Price (NOK)')
ax.set_title('EQNR.OL Stock Price Over Time')
plt.tight_layout()
plt.show()
```

```python
ax = data_close['returns'].plot(figsize=(12, 8))
ax.set_xlabel('Date')
ax.set_ylabel('Returns')
ax.set_title('EQNR.OL Daily Returns')
plt.tight_layout()
plt.show()
```
